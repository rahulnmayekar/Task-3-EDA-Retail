## Cleaning the Data

```python
#checking duplicate values

data.duplicated().sum()
```

```python
#dropping duplicate values

data.drop_duplicates(inplace=True) 
```

```python
retail_data = data.drop(['Country', 'City', 'Postal Code'], axis = 1)
```

#### It can be seen from the above observations that since the data is about one country 'United States', it can be dropped and also we can delete the 'City' and 'Postal Code' column as the information is not much useful to make analysis. 

```python
retail_data.head()
```

### Inclusion of new column - Net Sales and Cost

```python
# Net Sales
# Net sales = Gross sales - Discount

def net_sales(retail_data):
    net_sales = retail_data['Sales'] * (1 - retail_data['Discount'])
    return net_sales
retail_data['Net Sales'] = retail_data.apply(net_sales, axis = 1)
```

```python
retail_data.head()
```

```python
# Cost
# Profit/Loss = Net Sales - Cost 

def cost(retail_data):
    cost = retail_data['Net Sales'] - retail_data['Profit']
    return cost
retail_data['Cost'] = retail_data.apply(cost, axis = 1)
```

```python
retail_data.head()
```

#### The inclusion of 'cost' and 'net-sales' provides additional information to make use of data.

```python
# Looking for Outliers in Net Sales

sns.boxplot(x = retail_data['Net Sales'])
plt.show()
```

```python
# Removing the Outliers
retail_data.loc[retail_data['Net Sales'] > 10000, 'Net Sales'] = np.mean(retail_data['Net Sales'])

retail_data['Net Sales'].plot.box()
plt.show()
```

## Relationship Analysis

```python
corelation = retail_data.corr()
sns.heatmap(corelation, xticklabels = corelation.columns, yticklabels = corelation.columns, annot = True)
plt.show()
```

#### The above heatmap shows that the light shades represents positive correlation where as dark shades represents negative correlation.

```python
# The below scatter plot shows the distribution of various continuous variablels of the data.

sns.pairplot(retail_data)
plt.show()
```

```python
retail_data.groupby('Ship Mode')['Net Sales'].sum().plot.bar(color = 'red')
plt.show()
```

#### It can be seen that the maximum sales takes place in Standard Class shipment and minimum through Same Day shipment

```python
plt.figure(figsize = (10,6))
sns.countplot(x = 'Segment', hue = 'Ship Mode', data = retail_data)
plt.ylabel('Count of Ship Mode')
plt.show()
```

#### The ratio of Ship Mode and different Segment of customers is same which can be observed from the above graph.

```python
# Sales in various zones

retail_data.groupby('Segment')['Net Sales'].sum().plot.bar()
plt.show()
```

```python
# Segment Distribution on the basis of profit

retail_data.groupby('Segment')['Profit'].sum().plot.bar(color = 'Green')
plt.show()
```

#### From the above distribution it can be seen that the Net Sales and Profit in the segment are in proportion, so in order to increase the profit we more focus should be given to the net sales in Corporate and Home Office sector as they are generating low profit as compared to Consumer

```python
# Regional Distribution and Net Sales comparison alongwith profit

retail_data.groupby('Region')['Net Sales'].sum().plot(kind = 'pie', autopct = '%.1f%%', explode=[0.05]*4, pctdistance=0.5)
plt.show()
```

```python
retail_data.groupby('Region')['Profit'].sum().plot(kind = 'pie', autopct = '%.1f%%', explode=[0.05]*4, 
pctdistance=0.5, colors = ['lightblue', 'yellow', 'lightgreen', 'pink'])
plt.show()
```

#### The above data comprises of distribution of Net Sales and Profit in various regions of the country. The Profits from  East and West zones should be maintained as they generate higher profits in respect to their sales where as more focus must be given to Central and South zones especially central zone as the profit is considerably low as compared to sales in those zones

```python
# State-wise Distribution

retail_data.groupby('State')[['Net Sales', 'Profit']].sum().sort_values(by = 'Profit', 
ascending = True).plot(kind = 'bar', figsize = (16,7))
plt.show()
```

#### The states on the left side of the graph shown above are loss making areas and more attention should be given to them as eventhough the sales are in huge number in such areas, there is no profit visible.

```python
# Net Sales from different categories

retail_data.groupby('Category')['Net Sales'].sum().plot.bar(color = 'gold')
plt.show()
```

```python
# Profit making Categories

retail_data.groupby('Category')['Profit'].sum().plot.bar(color = 'lightgreen')
plt.show()
```

```python
# Scatter Plot of correlation of cost and net sales associated with product category

sns.relplot(x = 'Cost', y = 'Net Sales', hue = 'Category', data = retail_data)
plt.show()
```

```python
retail_data.groupby('Sub-Category')[['Net Sales', 'Profit']].sum().sort_values(by = 'Profit', 
ascending = True).plot(kind = 'barh', figsize = (16,7), color = ['Violet', 'purple'])
plt.show()
```

#### In the scatter plot above you can see the distribution of Net sales to their respective cost in relation to each product from Category also we can see cost associated with that of net sales is high in Furniture category leading to low profits as compared to other two categories. Tables a product from sub-category can be a reason of loss incurred from furniture category 

```python
# Category-Quantity-Discount Analysis
plt.figure(figsize = (15,8))
sns.lineplot(x = 'Discount', y = 'Profit', hue = 'Category', data = retail_data)
plt.show()
```

#### It can be observed from the above graph that there is no change in the profit of office supplies category respective of change in the discount where as technology sector incurrs significant loss when there is increase in the discount and furniture sector shows nominal change in value

```python
# Comparison of Net Sales and Profit

sns.kdeplot(data = retail_data['Net Sales'], color = 'red')
sns.kdeplot(data = retail_data['Profit'], color = 'blue')
plt.xlim(-500,1000)
plt.show()
```

```python
# Comparison of Net Sales and Cost

sns.kdeplot(data = retail_data['Net Sales'], color = 'green')
sns.kdeplot(data = retail_data['Cost'], color = 'orange')
plt.xlim(-500,2000)
plt.show()
```

* In the first graph we can see profits are more in comparison to Net Sales.
* In the second graph the Net Sales associated is in alignment with the cost.
